```python
# ==============================================================================
# Cell 2: The Refactored SimulationRun Class (with Data Truncation Logic)
# ==============================================================================
class SimulationRun:
    """
    A holistic class to load, process, and manage all data from a single HDF5 run.
    Data is loaded once and stored for multiple analyses.
    """
    def __init__(self, filename):
        self.filename = filename
        self.particle_centers = None
        self.num_particles = 0
        self.average_Z = -1
        self.Nv = -1
        self.is_bumpy = False
        self.cages = []
        self.raw_data = {}
        self.filtered_data = {}
        self.step_metadata = {}
        self.available_steps = []

        if not os.path.exists(filename):
            raise FileNotFoundError(f"File not found: {filename}")
        
        self._load_static_data()

    def _load_static_data(self):
        """Private method to load data that is constant for the whole run."""
        with h5py.File(self.filename, 'r') as f:
            self.average_Z = f.attrs.get('Z', -1)
            self.Nv = f.attrs.get('Nv', -1)
            self.is_bumpy = bool(f.attrs.get('bumpy', False))

            if '/vertices' in f:
                vertex_data = f['/vertices'][()]
                self.particle_centers = vertex_data[:, :2]
                self.num_particles = self.particle_centers.shape[0]
            if '/cages' in f:
                self.cages = [f['/cages'][name][()] for name in sorted(f['/cages'].keys())]
            
            self.available_steps = sorted([int(k) for k in f.keys() if k.isdigit()])

    def load_all_steps(self):
        """Loads and processes sample data from all available decompression steps."""
        print(f"Loading all steps for {os.path.basename(self.filename)}...")
        with h5py.File(self.filename, 'r') as f:
            for step_idx in self.available_steps:
                step_key = f"{step_idx:06d}"
                if step_key not in f: continue
                decompression_group = f[step_key]
                
                self.step_metadata[step_idx] = {
                    'dphi': decompression_group.attrs.get('dphi', 0.0),
                    'sigma': decompression_group.attrs.get('sigma', 0.0)
                }
                
                self.raw_data[step_idx] = {}
                for p_idx in range(self.num_particles):
                    particle_key = f"{p_idx:06d}"
                    if particle_key not in decompression_group: continue

                    samples = decompression_group[particle_key][()]
                    valid_mask = samples[:, 2] > 0
                    if not np.any(valid_mask): continue

                    valid_samples = samples[valid_mask]
                    delta_r = np.linalg.norm(valid_samples[:, :2] - self.particle_centers[p_idx], axis=1)

                    self.raw_data[step_idx][p_idx] = {
                        'dr': delta_r,
                        'dtheta': valid_samples[:, 2],
                        'samples_xy': valid_samples[:, :2]
                    }
        print("...loading complete.")

    def filter_all_data(self, epsilon=0.01):
        """
        Applies reachability filtering to all loaded raw data and stores the result.
        This is an expensive, one-time operation.
        """
        print(f"Filtering data for {os.path.basename(self.filename)}...")
        self.filtered_data = {}
        for step_idx, step_data in self.raw_data.items():
            self.filtered_data[step_idx] = {}
            for p_idx, p_data in step_data.items():
                center = self.particle_centers[p_idx]
                reachable_mask = get_reachable_mask(p_data['samples_xy'], center, epsilon)
                
                self.filtered_data[step_idx][p_idx] = {
                    'dr': p_data['dr'][reachable_mask],
                    'dtheta': p_data['dtheta'][reachable_mask],
                    'samples_xy': p_data['samples_xy'][reachable_mask]
                }
        print("...filtering complete.")

    def truncate_at_cage_break(self, epsilon=0.01):
        """
        Finds the first decompression step where any particle escapes its cage
        and removes that step and all subsequent steps from the run's data.
        """
        breaking_step = -1
        # Iterate through steps in order to find the *first* break
        for step_idx in self.available_steps:
            if step_idx not in self.filtered_data: continue

            for p_idx, p_data in self.filtered_data[step_idx].items():
                if p_idx >= len(self.cages): continue

                reachable_points = p_data['samples_xy']
                cage_vertices = self.cages[p_idx]

                if reachable_points.size == 0 or cage_vertices.size == 0: continue

                cage_tree = cKDTree(cage_vertices)
                min_dist, _ = cage_tree.query(reachable_points, k=1)

                if np.min(min_dist) < epsilon:
                    breaking_step = step_idx
                    break # Found a break for this step, no need to check other particles
            if breaking_step != -1:
                break # Found the first breaking step, exit the main loop
        
        if breaking_step != -1:
            dphi_at_break = self.step_metadata[breaking_step]['dphi'][0]
            print(f"  --> Cage break detected at step {breaking_step} (dphi={dphi_at_break:.2e}). Truncating data.")
            
            # Identify all steps from the breaking point onwards to remove
            steps_to_remove = [s for s in self.available_steps if s >= breaking_step]
            
            for step in steps_to_remove:
                self.raw_data.pop(step, None)
                self.filtered_data.pop(step, None)
                self.step_metadata.pop(step, None)
            
            # Update the list of available steps
            self.available_steps = [s for s in self.available_steps if s < breaking_step]
        else:
            print("  --> No cage breaking detected. Keeping all steps.")
```

```python
# Cell 3: Helper Functions

def get_reachable_mask(sample_points, center_point, epsilon=0.01):
    """
    Finds the contiguous cluster of points connected to a center point.
    Returns a boolean mask for the original sample_points array.
    """
    if sample_points.shape[0] == 0:
        return np.array([], dtype=bool)

    all_points = np.vstack([center_point, sample_points])
    tree = cKDTree(all_points)
    queue = [0]
    reachable_indices = {0}

    while queue:
        current_idx = queue.pop(0)
        neighbors_indices = tree.query_ball_point(all_points[current_idx], r=epsilon)
        for neighbor_idx in neighbors_indices:
            if neighbor_idx not in reachable_indices:
                reachable_indices.add(neighbor_idx)
                queue.append(neighbor_idx)

    mask = np.zeros(len(sample_points), dtype=bool)
    for idx in reachable_indices:
        if idx > 0:
            mask[idx - 1] = True
            
    return mask

def format_scientific_latex(n):
    """Formats a number into a LaTeX-compatible scientific notation string."""
    if np.abs(n - 0.0) <= 1e-9: return "0"
    mantissa, exponent = f"{n:.1e}".split('e')
    if (np.abs(float(mantissa) - 1.0) <= 1e-9):
        return fr"10^{{{int(exponent)}}}"
    else:
        return fr"{mantissa} \times 10^{{{int(exponent)}}}"
```

```python
# Cell 4: Data Loading and Pre-Processing

# --- Configuration ---
file_list = glob("../../runs/cage2/bumpy/*.h5")

# --- 1. Load All Static and Dynamic Data ---
runs = []
for f in sorted(file_list):
    try:
        run = SimulationRun(f)
        run.load_all_steps()
        runs.append(run)
    except (FileNotFoundError, KeyError, IndexError) as e:
        print(f"Could not process {f}: {e}")

print(f"\nSuccessfully loaded {len(runs)} simulation run(s).")

# --- 2. Perform Expensive Filtering (Once) ---
# Now that all data is in memory, apply the filter.
# The results are stored in the 'filtered_data' attribute of each object.
for run in runs:
    run.filter_all_data(epsilon=0.01)

# --- 3. Truncate Data at Cage-Breaking Point ---
print("\nChecking for cage breaking and truncating data...")
for run in runs:
    run.truncate_at_cage_break(epsilon=0.01)
```

```python
from scipy.spatial import cKDTree

# --- Configuration ---
run_to_plot = runs[1]
step_to_plot = 6
p_idx = 4  # particle to analyze
sigma = 0.00001  # smoothing scale

# --- Get relevant data ---
samples = run_to_plot.filtered_data[step_to_plot][p_idx]['samples_xy']
dthetas = run_to_plot.filtered_data[step_to_plot][p_idx]['dtheta']

# --- KD-tree neighborhood scoring ---
tree = cKDTree(samples)
scores = np.zeros(len(samples))

for i, (pt, dtheta_i) in enumerate(zip(samples, dthetas)):
    idxs = tree.query_ball_point(pt, r=0.01)
    neighbors = samples[idxs]
    neighbor_dthetas = dthetas[idxs]
    distances = np.linalg.norm(neighbors - pt, axis=1)
    weights = np.exp(-distances**2 / (2 * sigma**2))
    scores[i] = np.sum(weights * neighbor_dthetas)

# --- Peak detection ---
peak_idx = np.argmax(scores)
peak_point = samples[peak_idx]

# --- Plotting (reuse your previous heatmap code with an extra point) ---
fig, ax = plt.subplots(figsize=(10, 10))
valid_dthetas = dthetas[dthetas > 0]
vmin = valid_dthetas.min() if valid_dthetas.size > 0 else 1e-3
vmax = valid_dthetas.max() if valid_dthetas.size > 0 else 1.0

sc = ax.scatter(
    samples[:, 0], samples[:, 1], c=dthetas,
    cmap='plasma', s=7, edgecolors='none',
    norm=LogNorm(vmin=vmin, vmax=vmax)
)

# Existing overlay points
center = run_to_plot.particle_centers[p_idx]
cage_centroid = np.mean(run_to_plot.cages[p_idx], axis=0)
sample_centroid = np.mean(samples, axis=0)

ax.add_patch(Polygon(run_to_plot.cages[p_idx], closed=True, fill=False, edgecolor='black', linewidth=1.2))
ax.scatter(center[0], center[1], c='white', edgecolor='black', s=30, zorder=5)
ax.scatter(cage_centroid[0], cage_centroid[1], c='white', edgecolor='red', s=30, zorder=5)
ax.scatter(sample_centroid[0], sample_centroid[1], c='white', edgecolor='green', s=30, zorder=5)
ax.scatter(peak_point[0], peak_point[1], c='white', edgecolor='blue', s=30, zorder=5)

# Annotate
ax.annotate("Peak", xy=peak_point, xytext=(10, -10), textcoords="offset points",
            color='blue', fontweight='bold', fontsize=12,
            bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="blue", lw=0.5))
ax.annotate("Center", xy=center, xytext=(10, 10), textcoords="offset points",
            color='black', fontweight='bold', fontsize=12,
            bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="black", lw=0.5))
ax.annotate("Cage", xy=cage_centroid, xytext=(10, -10), textcoords="offset points",
            color='red', fontweight='bold', fontsize=12,
            bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="red", lw=0.5))
ax.annotate("Samples", xy=sample_centroid, xytext=(10, 10), textcoords="offset points",
            color='green', fontweight='bold', fontsize=12,
            bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="green", lw=0.5))

# Final formatting
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_aspect('equal', adjustable='box')
ax.set_xlim(0.7, 1.0)
ax.set_ylim(0.5, 0.8)
cbar = fig.colorbar(sc, ax=ax, shrink=0.7)
cbar.set_label(r"$\Delta\theta$")
plt.savefig("heatmap_four_points.svg")
plt.show()
```

```python
# Cell 8: Per-Particle ⟨Δθ⟩ vs Δr Curves for ALL Runs

# --- Configuration ---
# Use all runs loaded in Cell 4
runs_to_analyze = runs
# The decompression steps you want to visualize
steps_to_plot = [0, 4, 5, 6, 7, 8]
num_bins = 20 # Use fewer bins for cleaner individual curves

# --- Visualization Setup ---
fig, ax = plt.subplots()
colors = plt.cm.plasma(np.linspace(0, 1, len(steps_to_plot)))

# --- Main Analysis Loop ---
# Loop over each decompression step to define the color and legend entry
for i, step in enumerate(steps_to_plot):
    
    step_color = colors[i]
    # Get dphi from the first available run for the legend label
    dphi = next((run.step_metadata[step]['dphi'][0] for run in runs_to_analyze if step in run.step_metadata), 0.0)
    step_label = fr"$\Delta\phi = {format_scientific_latex(dphi)}$"

    # Now, loop through every run and every particle to plot their curves
    for run in runs_to_analyze:
        if step in run.filtered_data:
            for p_idx, particle_data in run.filtered_data[step].items():
                
                dr_vals = particle_data['dr']
                dtheta_vals = particle_data['dtheta']

                if dr_vals.size < 2: continue

                # --- Binning Logic for this specific particle ---
                linear_bins = np.linspace(dr_vals.min(), dr_vals.max(), num_bins + 1)
                theta_sum, bin_edges = np.histogram(dr_vals, bins=linear_bins, weights=dtheta_vals)
                counts, _ = np.histogram(dr_vals, bins=linear_bins)
                avg_dtheta = np.divide(theta_sum, counts, out=np.full(theta_sum.shape, np.nan), where=(counts > 0))
                bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

                # --- Plot this particle's curve ---
                # Use a low alpha to see the density of overlapping curves
                ax.plot(
                    bin_centers, 
                    avg_dtheta, 
                    linestyle='-', 
                    marker='o',
                    color=step_color,
                    alpha=0.5 # Make individual curves very faint
                )

    # After plotting all curves for this step, add one representative line for the legend
    ax.plot([], [], linestyle='-', color=step_color, label=step_label, lw=2)

# --- Formatting ---
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel(r"Radial Displacement, $\Delta r$")
ax.set_ylabel(r"Per-Particle Average, $\langle\Delta\theta\rangle$")
ax.set_title(r"Aggregated Particle Behavior Across All Runs")
legend = ax.legend(title="Decompression")
legend.get_frame().set_edgecolor('black')
plt.tight_layout()
plt.show()
```

```python
# ==============================================================================
# Cell 6: Visualization - System-Ensemble Averaged ⟨Δθ⟩ vs Δr
# ==============================================================================
# This cell plots one smoothed curve for each decompression step, averaged over
# ALL particles from ALL runs.

# --- Configuration ---
steps_to_plot = [0, 3, 4, 5, 6, 7]
num_bins = 20

# --- Visualization Setup ---
fig, ax = plt.subplots()
colors = plt.cm.plasma(np.linspace(0, 1, len(steps_to_plot)))

# --- Main Analysis Loop ---
for i, step in enumerate(steps_to_plot):
    # Aggregate data for this step from all runs and all particles
    all_dr_for_step = []
    all_dtheta_for_step = []
    
    for run in runs:
        if step in run.filtered_data:
            for p_idx, particle_data in run.filtered_data[step].items():
                all_dr_for_step.append(particle_data['dr'])
                all_dtheta_for_step.append(particle_data['dtheta'])

    # Check if we gathered any data for this step
    if not all_dr_for_step:
        print(f"No data found for step {step}. Skipping.")
        continue
        
    # Combine the lists of arrays into single large arrays
    dr_vals = np.concatenate(all_dr_for_step)
    dtheta_vals = np.concatenate(all_dtheta_for_step)
    
    if dr_vals.size == 0: continue

    # --- Binning Logic on the aggregated data ---
    linear_bins = np.linspace(dr_vals.min(), dr_vals.max(), num_bins + 1)
    theta_sum, bin_edges = np.histogram(dr_vals, bins=linear_bins, weights=dtheta_vals)
    counts, _ = np.histogram(dr_vals, bins=linear_bins)
    avg_dtheta = np.divide(theta_sum, counts, out=np.full(theta_sum.shape, np.nan), where=(counts > 0))
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    # --- Plotting ---
    dphi = next((run.step_metadata[step]['dphi'][0] for run in runs if step in run.step_metadata), 0.0)
    label = fr"$\Delta\phi = {format_scientific_latex(dphi)}$"
    
    ax.plot(
        bin_centers, avg_dtheta, marker='o', linestyle='-', color=colors[i], mew=1.5, label=label
    )

# --- Formatting ---
#ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlabel(r"$\Delta r$")
ax.set_ylabel(r"$\langle\Delta\theta\rangle$")
legend = ax.legend()
legend.get_frame().set_edgecolor('black')
plt.tight_layout()
plt.savefig("binned_radial.svg")
plt.show()
```

```python
# ==============================================================================
# Unified Plot: ⟨Δθ⟩ vs Δr using 4 center types with cleaner labeling
# ==============================================================================

from matplotlib.lines import Line2D

# --- Configuration ---
steps_to_plot = [0, 3, 4, 5, 6]
num_bins = 15
center_types = ['particle', 'cage', 'samples', 'peak']
linestyles = ['-', '--', ':', '-.']  # One per center type
markers = ['o', 's', '^', 'x']  # One per center type (circle, square, triangle, diamond)
colors = plt.cm.plasma(np.linspace(0, 1, len(steps_to_plot)))

# --- Visualization Setup ---
fig, ax = plt.subplots(figsize=(10, 7))

# --- Main Plotting Loop ---
for i, step in enumerate(steps_to_plot):
    for j, center_type in enumerate(center_types):
        all_dr = []
        all_dtheta = []

        for run in runs:
            if step not in run.filtered_data:
                continue

            for p_idx, pdata in run.filtered_data[step].items():
                if pdata['samples_xy'].shape[0] < 5:
                    continue

                # Determine center point
                if center_type == 'particle':
                    center = run.particle_centers[p_idx]
                elif center_type == 'cage':
                    center = np.mean(run.cages[p_idx], axis=0)
                elif center_type == 'samples':
                    center = np.mean(pdata['samples_xy'], axis=0)
                elif center_type == 'peak':
                    tree = cKDTree(pdata['samples_xy'])
                    scores = np.zeros(len(pdata['samples_xy']))
                    sigma = 0.00001
                    for x, (pt, dtheta_i) in enumerate(zip(pdata['samples_xy'], pdata['dtheta'])):
                        idxs = tree.query_ball_point(pt, r=0.01)
                        neighbors = pdata['samples_xy'][idxs]
                        neighbor_dthetas = pdata['dtheta'][idxs]
                        distances = np.linalg.norm(neighbors - pt, axis=1)
                        weights = np.exp(-distances**2 / (2 * sigma**2))
                        scores[x] = np.sum(weights * neighbor_dthetas)
                    
                    # --- Peak detection ---
                    peak_idx = np.argmax(scores)
                    center = pdata['samples_xy'][peak_idx]
                    
                delta_r = np.linalg.norm(pdata['samples_xy'] - center, axis=1)
                all_dr.append(delta_r)
                all_dtheta.append(pdata['dtheta'])

        if not all_dr:
            continue

        dr_vals = np.concatenate(all_dr)
        dtheta_vals = np.concatenate(all_dtheta)

        bins = np.linspace(dr_vals.min(), dr_vals.max(), num_bins + 1)
        theta_sum, bin_edges = np.histogram(dr_vals, bins=bins, weights=dtheta_vals)
        counts, _ = np.histogram(dr_vals, bins=bins)
        avg_dtheta = np.divide(theta_sum, counts, out=np.full(theta_sum.shape, np.nan), where=(counts > 0))
        bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

        dphi = next((run.step_metadata[step]['dphi'][0] for run in runs if step in run.step_metadata), 0.0)
        label = fr"$\Delta\phi = {format_scientific_latex(dphi)}$"

        ax.plot(bin_centers, avg_dtheta,
                linestyle=linestyles[j % len(linestyles)],
                marker=markers[j % len(markers)],
                color=colors[i],
                mew=1.0,
                markersize=4,
                label=label if j == 0 else None)  # Avoid repeating labels

# --- Axis Formatting ---
#ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlabel(r"$\Delta r$")
ax.set_ylabel(r"$\langle\Delta\theta\rangle$")

# --- Create a custom legend for center types ---
legend_elements_centers = [
    Line2D([0], [0], color='black', linestyle=linestyles[k], marker=markers[k],
           label=center_types[k].capitalize(), lw=1.5, markersize=6)
    for k in range(len(center_types))
]

first_legend = ax.legend(fontsize=9)
ax.add_artist(first_legend)
second_legend = ax.legend(handles=legend_elements_centers, loc='lower left', fontsize=9)

plt.tight_layout()
plt.savefig("binned_radial_centers.svg")
plt.show()
```

```python
# ==============================================================================
# Cell X: Distance-Based PDFs Between All 4 Points (6 Combinations) + Means
# ==============================================================================

steps_to_plot = [0, 3, 4, 5, 6]
num_steps = len(steps_to_plot)
colors = plt.cm.plasma(np.linspace(0, 1, num_steps))
markers = ['o', 'x', 's', 'd', '^']
linestyles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]
num_bins = 50
min_count_threshold = 1

distance_labels = [
    r"$|\mathbf{x}_{\mathrm{particle}} - \mathbf{x}_{\mathrm{cage}}|$",
    r"$|\mathbf{x}_{\mathrm{particle}} - \mathbf{x}_{\mathrm{samples}}|$",
    r"$|\mathbf{x}_{\mathrm{particle}} - \mathbf{x}_{\mathrm{peak}}|$",
    r"$|\mathbf{x}_{\mathrm{samples}} - \mathbf{x}_{\mathrm{cage}}|$",
    r"$|\mathbf{x}_{\mathrm{samples}} - \mathbf{x}_{\mathrm{peak}}|$",
    r"$|\mathbf{x}_{\mathrm{cage}} - \mathbf{x}_{\mathrm{peak}}|$"
]

fig, axs = plt.subplots(2, 3, figsize=(20, 10), sharey=True)
axs = axs.flatten()

stepwise_distances = {step: [[] for _ in range(6)] for step in steps_to_plot}
legend_handles = []
legend_labels = []

for run in runs:
    for step in steps_to_plot:
        if step not in run.filtered_data:
            continue

        for p_idx, pdata in run.filtered_data[step].items():
            if p_idx >= len(run.cages) or pdata['samples_xy'].shape[0] < 5:
                continue

            center = run.particle_centers[p_idx]
            cage = run.cages[p_idx]
            samples = pdata['samples_xy']
            dtheta = pdata['dtheta']

            cage_centroid = np.mean(cage, axis=0)
            sample_centroid = np.mean(samples, axis=0)

            tree = cKDTree(samples)
            scores = np.zeros(len(samples))
            sigma = 0.00001
            
            for x, pt in enumerate(samples):
                idxs = tree.query_ball_point(pt, r=0.01)
                if not idxs:
                    continue  # skip if no neighbors found
                neighbors = samples[idxs]
                neighbor_dthetas = dtheta[idxs]
                distances = np.linalg.norm(neighbors - pt, axis=1)
                weights = np.exp(-distances**2 / (2 * sigma**2))
                if weights.sum() == 0:
                    continue
                scores[x] = np.sum(weights * neighbor_dthetas) / weights.sum()
            
            # Check for non-empty scores before using argmax
            if len(scores) == 0 or np.all(scores == 0):
                continue  # avoid using uninitialized peak
            
            peak_idx = np.argmax(scores)
            if peak_idx >= len(samples):
                continue  # safety against overflow
            
            peak_point = samples[peak_idx]

            # Distances for 6 combinations
            distances = [
                np.linalg.norm(center - cage_centroid),
                np.linalg.norm(center - sample_centroid),
                np.linalg.norm(center - peak_point),
                np.linalg.norm(sample_centroid - cage_centroid),
                np.linalg.norm(sample_centroid - peak_point),
                np.linalg.norm(cage_centroid - peak_point),
            ]
            for i in range(6):
                stepwise_distances[step][i].append(distances[i])

# Plot each distance type on a subplot
for dist_idx, ax in enumerate(axs):
    for i, step in enumerate(steps_to_plot):
        data = np.array(stepwise_distances[step][dist_idx])
        data = data[data > 0]
        if len(data) < 5:
            continue

        bins = np.logspace(np.log10(data.min()), np.log10(data.max()), num_bins + 1)
        counts, _ = np.histogram(data, bins=bins)
        bin_widths = np.diff(bins)
        pdf = counts / (np.sum(counts) * bin_widths)
        bin_centers = np.sqrt(bins[:-1] * bins[1:])
        valid = counts >= min_count_threshold

        bin_centers = bin_centers[valid]
        pdf = pdf[valid]

        dphi = next((run.step_metadata[step]['dphi'][0] for run in runs if step in run.step_metadata), 0.0)
        label = fr"$\Delta\phi = {format_scientific_latex(dphi)}$"
        line, = ax.plot(
            bin_centers,
            pdf,
            marker=markers[i % len(markers)],
            linestyle=linestyles[i % len(linestyles)],
            color=colors[i],
            label=label
        )

        # --- Add vertical line for the mean ---
        mean_val = np.mean(data)
        ax.axvline(mean_val, color=colors[i], linestyle='--', linewidth=1.0, alpha=0.7)

        if dist_idx == 0:
            legend_handles.append(line)
            legend_labels.append(label)

    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_title(distance_labels[dist_idx])

axs[0].set_ylabel("Probability Density")
axs[3].set_ylabel("Probability Density")
axs[4].set_xlabel("Distance")

fig.legend(legend_handles, legend_labels, loc="center right", bbox_to_anchor=(1.1, 0.5))
plt.tight_layout()
plt.savefig("distance_pdfs.svg", bbox_inches='tight')
plt.show()
```

```python
# Cell 7: Analysis of Normalized Angular Freedom (Corrected)

# --- Configuration ---
run_to_analyze = runs[0]
steps_to_compare = [0, 1, 2, 3, 4, 5, 6, 7] 
num_bins = 50

# --- Data Preparation ---
if not run_to_analyze.process_step(decompression_step=0):
    print("Error: Could not process the jammed state (step 0). Cannot proceed.")
else:
    # Store the binned data AND THE BIN EDGES for the jammed state
    jammed_state_data = {}
    for p_idx, data in run_to_analyze.data_by_particle.items():
        # Define a common set of bin edges for this particle
        bin_edges = np.linspace(data['dr'].min(), data['dr'].max(), num_bins + 1)
        
        theta_sum, _ = np.histogram(data['dr'], bins=bin_edges, weights=data['dtheta'])
        counts, _ = np.histogram(data['dr'], bins=bin_edges)
        
        avg_dtheta_jammed = np.divide(theta_sum, counts, out=np.full_like(theta_sum, np.nan), where=(counts > 0))
        bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
        
        jammed_state_data[p_idx] = {
            'bin_centers': bin_centers,
            'avg_dtheta': avg_dtheta_jammed,
            'bin_edges': bin_edges  # <-- Store the edges for later use
        }

    # --- Main Analysis Loop ---
    master_curves = {}
    for step_idx in steps_to_compare:
        if not run_to_analyze.process_step(decompression_step=step_idx):
            continue

        particle_ratio_curves = []
        for p_idx, data in run_to_analyze.data_by_particle.items():
            if p_idx not in jammed_state_data: continue

            # FIX: Use the stored 'bin_edges' for a consistent histogram size
            bins_to_use = jammed_state_data[p_idx]['bin_edges']
            
            theta_sum, _ = np.histogram(data['dr'], bins=bins_to_use, weights=data['dtheta'])
            counts, _ = np.histogram(data['dr'], bins=bins_to_use)
            avg_dtheta_decompressed = np.divide(theta_sum, counts, out=np.full_like(theta_sum, np.nan), where=(counts > 0))

            # Now the division will work because both arrays have the same length
            ratio = np.divide(
                avg_dtheta_decompressed,
                jammed_state_data[p_idx]['avg_dtheta'],
                out=np.full_like(avg_dtheta_decompressed, np.nan),
                where=(jammed_state_data[p_idx]['avg_dtheta'] > 1e-9)
            )
            particle_ratio_curves.append(ratio)

        if particle_ratio_curves:
            master_curves[step_idx] = {
                'bin_centers': jammed_state_data[0]['bin_centers'],
                'avg_ratio': np.nanmean(np.array(particle_ratio_curves), axis=0)
            }

    # --- Plotting the Master Curves (Unchanged) ---
    fig, ax = plt.subplots()
    colors = plt.cm.plasma(np.linspace(0, 1, len(steps_to_compare)))

    for i, step_idx in enumerate(steps_to_compare):
        if step_idx in master_curves:
            curve = master_curves[step_idx]
            dphi = run_to_analyze.step_metadata[step_idx]['dphi']
            label = fr"$\Delta\phi = {dphi:.1e}$" if dphi > 0 else r"$\Delta\phi = 0$"
            
            ax.plot(
                curve['bin_centers'],
                curve['avg_ratio'],
                marker='o', linestyle='-', color=colors[i],
                fillstyle='none', mew=1.5, label=label
            )

    ax.set_xlabel(r"Radial Distance, $\Delta r$")
    ax.set_ylabel(r"Normalized Angular Freedom, $\langle\Delta\theta(\Delta\phi)\rangle / \langle\Delta\theta(0)\rangle$")
    ax.set_title(f"Evolution of Normalized Freedom\n(Run: {os.path.basename(run_to_analyze.filename)})")
    ax.axhline(1.0, color='black', linestyle='--', lw=1.5, label='Jammed State Baseline')
    ax.grid(True, which='both', linestyle=':', alpha=0.6)
    ax.set_xscale('log')

    legend = ax.legend(title=r"Decompression")
    legend.get_frame().set_edgecolor('black')
    plt.tight_layout()
    plt.show()
```
